```python
#setosaëŠ” ë¹¨ê°„ìƒ‰, versicolorëŠ” ë…¸ë€ìƒ‰, virginicaëŠ” íŒŒë€ìƒ‰

color=['r', 'y', 'b']


# setosaì˜ target ê°’ì€ 0, versicolorëŠ” 1, virginicaëŠ” 2. 

# ê° target ë³„ë¡œ ë‹¤ë¥¸ ìƒ‰ìœ¼ë¡œ scatter plot 

for i, c in enumerate(color):

    x_axis_data = df[df['target']==i]['sepal_length']
    y_axis_data = df[df['target']==i]['sepal_width']
    plt.scatter(x_axis_data, y_axis_data,color = c,label=iris.target_names[i])



plt.legend()

plt.xlabel('sepal length')

plt.ylabel('sepal width')

plt.show()
```

```python
# Target ê°’ì„ ì œì™¸í•œ ëª¨ë“  ì†ì„± ê°’ì„ MinMaxScalerë¥¼ ì´ìš©í•˜ì—¬ ë³€í™˜

# 'sepal_length','sepal_width','petal_length','petal_width'

df_features = df[['sepal_length','sepal_width','petal_length','petal_width']]

df_scaler = MinMaxScaler().fit_transform(df_features)



# PCAë¥¼ ì´ìš©í•˜ì—¬ 4ì°¨ì› ë³€ìˆ˜ë¥¼ 2ì°¨ì›ìœ¼ë¡œ ë³€í™˜



pca = PCA(n_components=2)



#fit( )ê³¼ transform( ) ì„ í˜¸ì¶œí•˜ì—¬ PCA ë³€í™˜ / ë°ì´í„° ë°˜í™˜

pca.fit(df_scaler)

df_pca = pca.transform(df_scaler)

print(df_pca.shape)
```

```python

# PCA ë³€í™˜ëœ ë°ì´í„°ì˜ ì»¬ëŸ¼ëª…ì„ ê°ê° PCA_1, PCA_2ë¡œ ì§€ì •



df_pca = pd.DataFrame(df_pca)

df_pca.columns = ['PCA_1','PCA_2']

df_pca['target']=df.target

df_pca.head(3)
```

```python
#setosaëŠ” ë¹¨ê°„ìƒ‰, versicolorëŠ” ë…¸ë€ìƒ‰, virginicaëŠ” íŒŒë€ìƒ‰

color=['r', 'y', 'b']



# setosaì˜ target ê°’ì€ 0, versicolorëŠ” 1, virginicaëŠ” 2. 

# ê° target ë³„ë¡œ ë‹¤ë¥¸ ìƒ‰ìœ¼ë¡œ scatter plot 

for i, c in enumerate(color):

    x_axis_data = df_pca[df_pca['target']==i]['PCA_1']

    y_axis_data = df_pca[df_pca['target']==i]['PCA_2']

    plt.scatter(x_axis_data, y_axis_data, color = c,label=iris.target_names[i])



plt.legend()

plt.xlabel('PCA_1')

plt.ylabel('PCA_2')

plt.show()
```
