```python
import pandas as pd

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
```

## 👋 실습
---
결측치가 있는 피쳐를 살펴본 후 대체

## 결측치가 있는 피쳐 살펴보기
---

```python
print(train.isnull().sum())
```

```python
print(test.isnull().sum())
```

```python
#각 차트의 평균치로 결측치를 채우는 방법
train.fillna({'hour_bef_temperature':int(train['hour_bef_temperature'].mean())}, inplace=True)
train.fillna({'hour_bef_precipitation':int(train['hour_bef_temperature'].mean())}, inplace=True)
train.fillna({'hour_bef_windspeed':int(train['hour_bef_temperature'].mean())}, inplace=True)
train.fillna({'hour_bef_humidity':int(train['hour_bef_temperature'].mean())}, inplace=True)
train.fillna({'hour_bef_visibility':int(train['hour_bef_temperature'].mean())}, inplace=True)
train.fillna({'hour_bef_ozone':int(train['hour_bef_temperature'].mean())}, inplace=True)
train.fillna({'hour_bef_pm10':int(train['hour_bef_temperature'].mean())}, inplace=True)
train.fillna({'hour_bef_pm2.5':int(train['hour_bef_temperature'].mean())}, inplace=True)

#결측치 위의 수치와 아래 수치간 평균으로 결측치를 채우는 방법
train.interpolate(inplace=True)

```
