1. 러닝 레이트가 크면 경사면 내려가다가 저점 주변에서 왔다갔다한다. 너무크면 오히려 점점 올라가서 바깥으로 튕겨나(발산)버린다. 반대로 너무 작으면 저점에 갈때까지 한참걸린다.

그렇다면 러닝레이트를 어떻게 정하냐? 특별한방법은 없다. 일단 0.01로 했다가 돌려보면서 바꿔라.

2. Data preprocessing for gredient descent

데이터중 하나가 뜬금없이 튀면 어떻게 처리해야 하나?

Standatdization, min-max scale 이 있다.

Standatdization : X_trd[:,0] = (X[:,0].mean()) / X[:,0].std()

min-max scale : xy = MinMaxScaler(xy)

3. Overfitting

모델이 학습용 데이터에 너무 잘맞으면 오히려 실전에선 안맞는 경우가 생긴다.

오버피팅을 줄이려면 어떡해야하나?

1. 더 많은 학습용 데이터

2. 피쳐의 개수(출력의 가짓수)를 줄여본다

3. Regulatization(일반화)

일반화 : W값이 너무크면 안된다.(왜안되는지는 모르겠다)
그러니 W가 너무 크면

l2reg = 0.001 * tf.reduce_sum(tf.square(W)) 하면 된다.
