```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
```

```python
train = pd.read_csv('data/train.csv') 
test = pd.read_csv('data/test.csv')
```

## ì „ì²˜ë¦¬
---
train ì€ interpolate() ë¥¼ ì´ìš©í•´ ê²°ì¸¡ì¹˜ë¥¼ DataFrame ê°’ì— ì„ í˜•ìœ¼ë¡œ ë¹„ë¡€í•˜ì—¬ ë³´ê°„í•©ë‹ˆë‹¤.
test ì˜ ê²°ì¸¡ì¹˜ëŠ” 0ìœ¼ë¡œ ì±„ì›Œì¤ë‹ˆë‹¤.

```python
train.interpolate(inplace=True)
test.fillna(0, inplace=True)
```

```python
# ì• ê³¼ì •ì—ì„œ íŒŒì•…ëœ ë³€ìˆ˜ ì¤‘ìš”ë„ê°€ ë‚®ì€ í”¼ì³ë“œë¥¼ ì œê±°í•©ë‹ˆë‹¤.

X_train = train.drop(['count', 'id', 'hour_bef_windspeed', 'hour_bef_pm2.5'], axis=1)
Y_train = train['count']
test = test.drop(['id', 'hour_bef_windspeed', 'hour_bef_pm2.5'], axis=1)
```

## ğŸ‘‹ ì‹¤ìŠµ
---

```python
#í•˜ì´í¼íŒŒë¼ë¯¸í„° ê°’ì€ ì‚¬ìš©ìê°€ ì§ì ‘ ì§€ì • í•´ì£¼ë©° ë¶„ì„ê°€ ë³¸ì¸ì˜ ê²½í—˜ì— ë”°ë¼ ì„ ì •í•˜ëŠ” ê²½ìš°ê°€ ë§ìŠµë‹ˆë‹¤.

from sklearn.model_selection import GridSearchCV

model = RandomForestRegressor(criterion = 'mse', random_state=2020)

params = {'n_estimators': [200, 300, 500],
          'max_features': [5, 6, 8],
          'min_samples_leaf': [1, 3, 5]}

greedy_CV = GridSearchCV(model, param_grid=params, cv = 3, n_jobs = -1)
greedy_CV.fit(X_train, Y_train)
```

```python
pred = greedy_CV.predict(test)
```
